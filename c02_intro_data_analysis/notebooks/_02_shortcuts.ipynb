```python
import random

import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

# Define a list of 10 customer names
customer_names = [
    'Alice Johnson', 'Bob Smith', 'Charlie Brown', 'Diana Prince',
    'Ethan Hunt', 'Fiona Clarke', 'George Miller', 'Hanna Davis',
    'Ian Wright', 'Jessica Lee',
]

# Define a list of sample purchase items
purchase_items_list = [
    'Laptop', 'Mouse', 'Smartphone', 'Headphones', 'Keyboard',
    'Tablet', 'Camera', 'Tripod', 'Monitor', 'Printer',
    'Smartwatch', 'Speakers', 'External Hard Drive', 'Router', 'Charger',
]

# Generate a dataset with 100 random purchases
expanded_data = {
    'Customer Name': [random.choice(customer_names) for _ in range(100)],
    'Purchase Items': [', '.join(random.sample(purchase_items_list, random.randint(1, 3))) for _ in range(100)],
    'Purchase Cost': [round(random.uniform(50, 1500), 2) for _ in range(100)],
}

# Convert to a DataFrame
customer_dataset = pd.DataFrame(expanded_data)
customer_dataset

# Show first 10 elements
customer_dataset.head(10)

# Describe dataset
customer_dataset.describe()

# Grouped data set
grouped_dataset = customer_dataset.groupby('Customer Name').sum().reset_index()
grouped_dataset

%matplotlib inline
plt.figure(figsize=(15, 5))
matplotlib.pyplot.bar(x=grouped_dataset['Customer Name'], height=grouped_dataset['Purchase Cost'], data=grouped_dataset)
```
