### CONJUNTO DE DATOS DE SERIES TEMPORALES

En esencia, un conjunto de datos de series temporales es una colección de puntos de
datos indexados en orden cronológico, donde cada punto de datos está asociado a una
marca de tiempo específica.

Se suelen descomponer en 4 elementos:
- Tendencia (trend)
- Estacionalidad (seasonality)
- Ciclicidad
- Ruido (noise)

Es como una narración histórica, una crónica de la evolución de un fenómeno, ya sea el precio de una acción,
la temperatura de una ciudad o los latidos del corazón de un paciente.

El poder de los conjuntos de datos de series temporales reside en su capacidad para revelar tendencias,
patrones y estacionalidad en los datos. Analizando la trayectoria histórica de una variable,
se puede conocer su comportamiento pasado, predecir sus valores futuros y detectar anomalías o
valores atípicos que se desvían de la norma.

### CONJUNTOS DE DATOS TRANVERSALES (Instantana en el tiempo)

A diferencia de los conjuntos de datos de series temporales, que siguen los cambios a lo largo del tiempo, los conjuntos de datos
transversales ofrecen una instantánea de las relaciones entre diversas variables en un instante concreto. Permiten comparar y
contrastar diferentes grupos o individuos en función de sus atributos y comportamientos.

Los conjuntos de datos transversales son muy valiosos para explorar correlaciones, identificar patrones y probar hipótesis sobre
las relaciones entre distintas variables.

### CONJUNTOS DE DATOS DE PANEL (Lente Longitudinal)
Imagina que haces un seguimiento del rendimiento académico de un grupo de estudiantes a lo largo de varios años, recopilando datos sobre sus notas,
asistencia y actividades extraescolares en diferentes momentos. Este conjunto de datos, que sigue a los mismos individuos o entidades a lo largo
de múltiples periodos de tiempo, se conoce como conjunto de datos de panel o conjunto de datos longitudinales.

Combina la dimensión temporal de los datos de series temporales con la perspectiva transversal, ofreciendo una lente única para estudiar cómo
cambian los individuos o las entidades a lo largo del tiempo y cómo influyen diversos factores en estos cambios.

Los conjuntos de datos de panel son indispensables para comprender cómo cambian los individuos o los grupos a lo largo del tiempo y cómo
influyen en esos cambios diversos factores. Al hacer un seguimiento de los mismos sujetos a lo largo de múltiples periodos de tiempo,
los datos de panel nos permiten ver el impacto directo de variables específicas, como nuevas políticas o intervenciones, en el
comportamiento o los resultados de los sujetos.
